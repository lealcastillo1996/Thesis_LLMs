#Function definition

```python
from logging import exception
#Function to seach in internet
def search_internet(query):
  #Seach in google top links related to the query
  # to search
  query_context = ""
  query_input = query
  query = query_context + query_input
  try:
    links = []
    for j in search(query,3):
        links.append(j)
    #Appending final text
    final_text = ""
    for link in links:
      try:
        downloaded = trafilatura.fetch_url(link)
        final_text= final_text + "/n " + trafilatura.extract(downloaded)
      except:
        pass
  except Exception as e:
    print(e)
    final_text = ""
  return final_text

#Functionn to create chunks
def split_string(text,chunk_size,overlap):
    # Initialize an empty list to store the chunks
    chunks = []
    # Split the text into chunks of size chunk_size with an overlap of overlap
    for i in range(0, len(text), chunk_size - overlap):
        chunk = text[i:i + chunk_size]
        chunks.append(chunk)
    return chunks

# Function to preprocess text
def preprocess_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Convert to lowercase
    text = text.lower()
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = text.split()
    words = [word for word in words if word not in stop_words]
    # Join words into a single string
    text = ' '.join(words)
    return text

#Function to obtain context with tf-idf
def obtain_context_tf_idf(query,N,documents):
  # Preprocess documents
  processed_documents = [preprocess_text(doc) for doc in documents]
  # Create TF-IDF vectorizer
  vectorizer = TfidfVectorizer()
  # Fit the vectorizer to the processed documents
  tfidf_matrix = vectorizer.fit_transform(processed_documents)
  # Vectorize the query
  processed_query = preprocess_text(query)
  query_vector = vectorizer.transform([processed_query])
  # Calculate cosine similarity between each document and the query
  from sklearn.metrics.pairwise import cosine_similarity
  similarity_scores = cosine_similarity(tfidf_matrix, query_vector)
  # Get the top N most similar documents
  top_documents = similarity_scores.argsort(axis=0)[-N:].flatten()
  top_documents = np.flip(top_documents)
  context = ""
  # Print the top N most similar documents
  for doc_idx in top_documents:
      context = context + documents[doc_idx]
  doc = Document(page_content=context)
  doc_context = []
  doc_context.append(doc)
  text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
  docs = text_splitter.split_documents(doc_context)
  return (docs, context)

#Function to obtain context with sentence embeddings
def obtain_context_emb(queries,N,corpus):
  processed_corpus = [preprocess_text(cor) for cor in corpus]
  corpus_embeddings = embedder.encode(processed_corpus, convert_to_tensor=True)
  print(type(corpus_embeddings))
  # Query sentences:
  queries = [queries]
  # Find the closest 5 sentences of the corpus for each query sentence based on cosine similarity
  top_k = min(N, len(corpus))
  for query in queries:
      query_embedding = embedder.encode(preprocess_text(query), convert_to_tensor=True)
      # We use cosine-similarity and torch.topk to find the highest 5 scores
      cos_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]
      top_results = torch.topk(cos_scores, k=top_k)
      context = ""
      for score, idx in zip(top_results[0], top_results[1]):
          context = context + corpus[idx]
  doc = Document(page_content=context)
  doc_context = []
  doc_context.append(doc)
  text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
  docs = text_splitter.split_documents(doc_context)
  return (docs, context)

#Roberta with no search and tf idf
def roberta_nosearch_tf_idf(query,chunk_size,overlap,n_context,prompt = None):
    #Reading the info from disk
    # Open the file for reading
    with open('documentation.txt', 'r', encoding="utf-8") as file:
      # Read the contents of the file
      contents = file.read()
    #Cleaning text
    contents = contents.replace('\n', '')
    contents = contents.replace('/n', '')
    contents = contents + aditional_context
    #Split data into chunks
    chunks = split_string(contents,chunk_size,overlap)
    #Searching best context with tf-idf
    context_query = obtain_context_tf_idf(query,n_context,chunks)[1]
    # select which embeddings we want to use
    # a) Get predictions
    nlp = pipeline('question-answering', model = model, tokenizer= tokenizer)
    QA_input = {
    'question': query,
    'context': context_query
      }
    result = nlp(QA_input)
    result = result["answer"]
    result = result.replace('\n', '')
    result = result.replace('/n', '')
    return result




#Roberta with no search and tf idf
def roberta_nosearch_emb(query,chunk_size,overlap,n_context,prompt = None,):
    #Reading the info from disk
    # Open the file for reading
    with open('documentation.txt', 'r', encoding="utf-8") as file:
      # Read the contents of the file
      contents = file.read()
    #Cleaning text
    #contents = contents + final_text
    contents = contents.replace('\n', '')
    contents = contents.replace('/n', '')
    contents = contents + aditional_context
    #Split data into chunks
    chunks = split_string(contents,chunk_size,overlap)
    #Searching best context with tf-idf
    context_query = obtain_context_emb(query,n_context,chunks)[1]
    # select which embeddings we want to use
    # a) Get predictions
    nlp = pipeline('question-answering', model = model, tokenizer= tokenizer)
    QA_input = {
    'question': query,
    'context': context_query
      }
    result = nlp(QA_input)
    result = result["answer"]
    result = result.replace('\n', '')
    result = result.replace('/n', '')
    return result


#Roberta with search and tf idf
def roberta_search_tf_idf(query,chunk_size,overlap,n_context,prompt = None):
    #Search in internet
    internet_text= search_internet(query)
    #Reading the info from disk
    # Open the file for reading
    with open('documentation.txt', 'r', encoding="utf-8") as file:
      # Read the contents of the file
      contents = file.read()
    #Cleaning text
    contents = contents + internet_text
    contents = contents.replace('\n', '')
    contents = contents.replace('/n', '')
    contents = contents + aditional_context
    #Split data into chunks
    chunks = split_string(contents,chunk_size,overlap)
    #Searching best context with tf-idf
    context_query = obtain_context_tf_idf(query,n_context,chunks)[1]
    # select which embeddings we want to use
    # a) Get predictions
    nlp = pipeline('question-answering', model = model, tokenizer= tokenizer)
    QA_input = {
    'question': query,
    'context': context_query
      }
    result = nlp(QA_input)
    result = result["answer"]
    result = result.replace('\n', '')
    result = result.replace('/n', '')
    return result

#Roberta with search and emn
def roberta_search_emb(query,chunk_size,overlap,n_context,prompt = None):

   #Search in internet
    internet_text= search_internet(query)
    #Reading the info from disk
    # Open the file for reading
    with open('documentation.txt', 'r', encoding="utf-8") as file:
      # Read the contents of the file
      contents = file.read()
    #Cleaning text
    contents = contents + internet_text
    contents = contents.replace('\n', '')
    contents = contents.replace('/n', '')
    contents = contents + aditional_context
    #Split data into chunks
    chunks = split_string(contents,chunk_size,overlap)
    #Searching best context with tf-idf
    context_query = obtain_context_emb(query,n_context,chunks)[1]
    # select which embeddings we want to use
    # a) Get predictions
    nlp = pipeline('question-answering', model = model, tokenizer= tokenizer)
    QA_input = {
    'question': query,
    'context': context_query
      }
    result = nlp(QA_input)
    result = result["answer"]
    result = result.replace('\n', '')
    result = result.replace('/n', '')
    return result
```
